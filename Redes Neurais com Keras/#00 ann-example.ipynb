#IMPORTANDO BIBLIOTECAS E PREPARANDO DATASET:

```python
import keras
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
x = pd.DataFrame(iris.data, columns=[iris.feature_names])
y = pd.Series(iris.target)
```

```python
x.head()
```

```python
y.value_counts()
```

```python
import keras
from keras.utils import np_utils
y_convertido = np_utils.to_categorical(y)
```

#PREPRANDO MINHA REDE

```python
y_convertido
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y_convertido, test_size = 0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
modelo = Sequential()
modelo.add(Dense(10, input_dim=4, kernel_initializer='normal', activation='relu'))
modelo.add(Dense(3, kernel_initializer='normal', activation='softmax'))
```

```python
otimizador = SGD()
```

```python
modelo.compile(loss='categorical_crossentropy', optimizer=otimizador, metrics=['acc'])
```

```python
#verbode 1 imprimiria as informções de cada época
modelo.fit(x_treino, y_treino, epochs=1000, batch_size=105, validation_data=(x_teste, y_teste), verbose=1) 
```

```python
predicoes = modelo.predict(x_teste)
print predicoes
```
