```python
#Inversão horizontal
import numpy as np
import cv2

for foto in gb.glob('/home/george/Área de Trabalho/anotacao natanael/inversao horizontal/*.jpg'):
    #print foto
    imagem = cv2.imread(foto)
    inverter = cv2.flip(imagem, 1)
    cv2.imwrite(foto, inverter) 
    #newsize = (500, 500) 
    #im = imRGB.resize(newsize)
    #im.save(foto)
```

```python
#Inversão vertical

for foto in gb.glob('/home/george/Área de Trabalho/anotacao natanael/inversao vertical/*.jpg'):
    #print foto
    imagem = cv2.imread(foto)
    inverter = cv2.flip(imagem, 0)
    cv2.imwrite(foto, inverter) 
```

```python
#rotacao 90, 180 e 270
import numpy as np
import cv2

for foto in gb.glob('/home/george/Área de Trabalho/anotacao natanael/90/*.jpg'):
    #print foto
    imagem = cv2.imread(foto)
    
    altura, largura = imagem.shape[:2]
    ponto = (largura / 2, altura / 2) #ponto no centro da figura
    rotacao = cv2.getRotationMatrix2D(ponto, 90, 1.0)
    rotacionado = cv2.warpAffine(imagem, rotacao, (largura, altura))
    

    cv2.imwrite(foto, rotacionado) 
```
